```python
import random

# Função obrigatória para formatar os valores conforme a regra de notas
def formatar_valor(valor):
    abs_val = abs(valor)
    if abs_val >= 1e5 or (abs_val <= 1e-4 and abs_val != 0):
        return f"{valor:.2e}"
    elif abs_val < 1.0:
        return f"{valor:.3f}"
    else:
        return f"{valor:.2f}"

print("=== INÍCIO DO PROGRAMA 1 (MOTOR A GASOLINA) ===")

# 1. Entrada de dados
num_cilindros = int(input("Digite a quantidade de cilindros do motor: "))

# 2. Dados Iniciais
P1 = 2175566.10
P2 = 1.45e-8
V1 = 12.2047
V2 = 27.4607
print(f"\n[Valores Iniciais]")
print(f"Pressão 1: {formatar_valor(P1)} psi")
print(f"Pressão 2: {formatar_valor(P2)} psi")
print(f"Volume 1: {formatar_valor(V1)} pol³")
print(f"Volume 2: {formatar_valor(V2)} pol³")

# 3. Processo Iterativo e Aleatório para encontrar o coeficiente n
# Relação: P1/P2 = (V2/V1)^n
alvo = P1 / P2
base = V2 / V1

limite_inferior = 0.0
limite_superior = 100.0
n_calculado = 0.0
tolerancia = alvo * 0.0001

print("\n[Iniciando processo iterativo aleatório para encontrar o coeficiente politrópico N]")
tentativas = 0

while True:
    tentativas += 1
    # Geração de valor aleatório dentro do limite convergente
    n_calculado = random.uniform(limite_inferior, limite_superior)
    resultado_teste = base ** n_calculado

    # Mostrando o processo na tela
    if tentativas % 10 == 0 or tentativas == 1:
        print(f"Tentativa {tentativas}: Testando n = {formatar_valor(n_calculado)}")

    if abs(resultado_teste - alvo) <= tolerancia:
        print(f"Convergência atingida na tentativa {tentativas}! N encontrado: {formatar_valor(n_calculado)}")
        break
    elif resultado_teste < alvo:
        limite_inferior = n_calculado
    else:
        limite_superior = n_calculado

# 4. Memória de Conversões Utilizadas
fator_lbf_pol_para_J = 0.11298 # (4.44822 N * 0.0254 m)
fator_J_para_CV = 735.50
fator_BTU_para_J = 1055.06
energia_gasolina_1L = 29890.36 * fator_BTU_para_J

print("\n[Conversões de Unidades Utilizadas no Sistema]")
print(f"1 lbf*pol equivale a {formatar_valor(fator_lbf_pol_para_J)} Joules")
print(f"1 CV equivale a {formatar_valor(fator_J_para_CV)} J/s")
print(f"Energia de 1 Litro de gasolina equivale a {formatar_valor(energia_gasolina_1L)} Joules")

# 5. Cálculos Termodinâmicos
trabalho_lbf_pol = ((P2 * V2) - (P1 * V1)) / (1.0 - n_calculado)
trabalho_J = trabalho_lbf_pol * fator_lbf_pol_para_J
potencia_1_cilindro_Js = trabalho_J / 60.0
potencia_total_Js = potencia_1_cilindro_Js * num_cilindros
potencia_total_CV = potencia_total_Js / fator_J_para_CV

tempo_segundos = 4.0 * 3600.0
energia_total_exigida_J = potencia_total_Js * tempo_segundos
consumo_litros = energia_total_exigida_J / energia_gasolina_1L

print("\n[Resultados Finais]")
print(f"Potência total do motor ({num_cilindros} cilindros): {formatar_valor(potencia_total_CV)} CV")
print(f"Consumo de combustível em 4 horas: {formatar_valor(consumo_litros)} Litros")
print("===============================================\n")
```
