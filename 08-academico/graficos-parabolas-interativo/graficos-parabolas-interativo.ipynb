```python
import matplotlib.pyplot as plt
import numpy as np

# Lista para armazenar os resultados
resultados = []
contador = 1

while True:
    a = float(input(f"Digite o valor de a para a parábola {contador} (ou 0 para sair): "))
    if a == 0:
        break
    b = float(input("Digite o valor de b: "))
    c = float(input("Digite o valor de c: "))

    x = np.linspace(0, 80, 400) # Intervalo de x entre 0 e 80 kg/h
    y = a*x**2 + b*x + c

    # Exibindo os 10 primeiros resultados
    print(f"\nCoordenadas da parábola {contador}:")
    for i in range(10):
        print(f"({x[i]:.2f}, {y[i]:.2f})")

    resultados.append((x, y))
    contador += 1

# Plotar todas as parábolas com rótulos
for i, (x, y) in enumerate(resultados):
    plt.plot(x, y, label=f"Parábola {i+1}")

plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.title('Gráfico das Parábolas')
plt.grid(True)
plt.legend()
plt.show()
```
