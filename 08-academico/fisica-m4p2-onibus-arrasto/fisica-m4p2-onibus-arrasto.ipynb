```python
# Função obrigatória para formatar os valores conforme a regra de notas
def formatar_valor(valor):
    abs_val = abs(valor)
    if abs_val >= 1e5 or (abs_val <= 1e-4 and abs_val != 0):
        return f"{valor:.2e}"
    elif abs_val < 1.0:
        return f"{valor:.3f}"
    else:
        return f"{valor:.2f}"

print("=== INÍCIO DO PROGRAMA 2 (ÔNIBUS DOUBLE-DECKER) ===")

# 1. Entrada de dados
velocidade_kmh = float(input("Digite a velocidade do ônibus (em km/h): "))
area_frontal = float(input("Digite a área frontal do ônibus (em m²): "))

# 2. Dados Constantes Iniciais (do módulo anterior)
coeficiente_arrasto = 0.40
massa_especifica_ar = 1.20
distancia_km = 584.0
potencia_ac_Js = 18400.0
potencia_atritos_Js = 47840.0
fator_J_para_CV = 735.50
energia_diesel_1L = 36017.05 * 1055.06 # BTU para Joules

# 3. Cálculos Mecânicos
velocidade_ms = velocidade_kmh / 3.6
forca_arrasto_N = 0.5 * coeficiente_arrasto * area_frontal * massa_especifica_ar * (velocidade_ms ** 2)
potencia_arrasto_Js = forca_arrasto_N * velocidade_ms

# 4. Conversão para CV e Potências Combinadas
potencia_arrasto_CV = potencia_arrasto_Js / fator_J_para_CV
potencia_atritos_CV = potencia_atritos_Js / fator_J_para_CV
perdas_onibus_CV = potencia_arrasto_CV + potencia_atritos_CV

potencia_ac_CV = potencia_ac_Js / fator_J_para_CV

potencia_combinada_Js = potencia_arrasto_Js + potencia_ac_Js + potencia_atritos_Js
potencia_combinada_CV = potencia_combinada_Js / fator_J_para_CV

# A potência do motor precisa ser exatamente igual à potência combinada exigida para manter o movimento
potencia_motor_CV = potencia_combinada_CV

# 5. Cálculo de Consumo
tempo_viagem_segundos = (distancia_km / velocidade_kmh) * 3600.0
energia_total_viagem_J = potencia_combinada_Js * tempo_viagem_segundos
consumo_litros = energia_total_viagem_J / energia_diesel_1L

print("\n[Resultados Finais - Potências e Consumo]")
print(f"Perdas do ônibus (Arrasto aerodinâmico + Atritos mecânicos): {formatar_valor(perdas_onibus_CV)} CV")
print(f"Potência exigida pelo ar-condicionado: {formatar_valor(potencia_ac_CV)} CV")
print(f"Todas as potências combinadas (Demanda total): {formatar_valor(potencia_combinada_CV)} CV")
print(f"Potência necessária fornecida pelo motor: {formatar_valor(potencia_motor_CV)} CV")
print(f"Consumo de combustível na viagem de 584 km: {formatar_valor(consumo_litros)} Litros")
print("===================================================\n")
```
