```python
import matplotlib.pyplot as plt
import numpy as np

# Configuração da figura
fig, ax = plt.subplots(figsize=(12, 8))
ax.set_xlim(-1, 7)
ax.set_ylim(-2, 5)
ax.set_aspect('equal')
ax.grid(True, linestyle='--', alpha=0.7)
ax.set_title('Diagrama de Corpo Livre (DCL) - Exercício 4 Corrigido', fontsize=16, fontweight='bold')

# Desenhar a barra (horizontal de 0 a 5 m)
barra_x = [0, 5]
barra_y = [0, 0]
ax.plot(barra_x, barra_y, 'k-', linewidth=4, label='Barra')

# Pontos de apoio
ax.plot(0, 0, 'ro', markersize=12, label='Apoio A (Engaste)')  # Ponto A - engaste na parede
ax.plot(5, 0, 'go', markersize=10, label='Ponto B')  # Ponto B - conexão com corda

# Força de 500 N no MEIO da barra (x = 2.5 m)
force_x = 2.5  # Meio da barra
force_angle = np.radians(0)  # Supondo força vertical para baixo
force_length = 1.0
dx = force_length * np.cos(force_angle)
dy = force_length * np.sin(force_angle)

ax.arrow(force_x, 0, 0, -1.2, head_width=0.2, head_length=0.2,
         fc='blue', ec='blue', linewidth=3)
ax.text(force_x + 0.3, -0.6, '500 N', fontsize=14, color='blue',
        weight='bold', ha='left')

# Corda no ponto B (triângulo 4-5-3)
corda_angle = np.arctan2(3, 4)  # Ângulo do triângulo 4-5-3
corda_length = 2.0
dx_corda = corda_length * np.cos(corda_angle)
dy_corda = corda_length * np.sin(corda_angle)

ax.arrow(5, 0, dx_corda, dy_corda, head_width=0.15, head_length=0.15,
         fc='purple', ec='purple', linewidth=3, linestyle='-')
ax.text(5 + dx_corda/2, dy_corda/2 + 0.3, 'Corda', fontsize=12, color='purple',
        weight='bold', ha='center')

# Indicar triângulo 4-5-3 na corda
ax.text(5 + dx_corda/3, dy_corda/4, '3', fontsize=11, color='purple',
        ha='center', bbox=dict(facecolor='white', alpha=0.7))
ax.text(5 + dx_corda/2, dy_corda/3, '4', fontsize=11, color='purple',
        ha='center', bbox=dict(facecolor='white', alpha=0.7))
ax.text(5 + dx_corda/2.5, dy_corda/2, '5', fontsize=11, color='purple',
        ha='center', bbox=dict(facecolor='white', alpha=0.7))

# Reações no apoio A (engaste)
reaction_length = 0.8
# Reação vertical em A
ax.arrow(0, 0, 0, reaction_length, head_width=0.15, head_length=0.15,
         fc='red', ec='red', linewidth=2)
ax.text(-0.3, reaction_length/2, 'R_Ay', fontsize=12, color='red',
        weight='bold', ha='center')

# Reação horizontal em A
ax.arrow(0, 0, -reaction_length, 0, head_width=0.15, head_length=0.15,
         fc='red', ec='red', linewidth=2)
ax.text(-reaction_length/2, 0.3, 'R_Ax', fontsize=12, color='red',
        weight='bold', ha='center')

# Momento em A
arc = plt.Circle((0, 0), 0.3, color='orange', fill=False, linewidth=2)
ax.add_patch(arc)
ax.text(0.4, 0.4, 'M_A', fontsize=12, color='orange',
        weight='bold', ha='center')

# Tração na corda (em B)
ax.arrow(5, 0, 0, reaction_length, head_width=0.15, head_length=0.15,
         fc='green', ec='green', linewidth=2)
ax.text(5.3, reaction_length/2, 'T', fontsize=12, color='green',
        weight='bold', ha='center')

# Indicar dimensões
ax.text(1.5, -1.0, '3 m', fontsize=12, ha='center', weight='bold',
        bbox=dict(facecolor='yellow', alpha=0.3))
ax.text(3.5, -1.0, '2 m', fontsize=12, ha='center', weight='bold',
        bbox=dict(facecolor='yellow', alpha=0.3))

# Linhas de dimensão
ax.plot([0, 0], [-0.7, -0.9], 'k-', linewidth=1)
ax.plot([3, 3], [-0.7, -0.9], 'k-', linewidth=1)
ax.plot([5, 5], [-0.7, -0.9], 'k-', linewidth=1)
ax.plot([0, 3], [-0.8, -0.8], 'k-', linewidth=1)
ax.plot([3, 5], [-0.8, -0.8], 'k-', linewidth=1)

# Legendas
legend_elements = [
    plt.Line2D([0], [0], color='blue', linewidth=3, label='Força de 500 N'),
    plt.Line2D([0], [0], color='red', linewidth=2, label='Reações em A'),
    plt.Line2D([0], [0], color='green', linewidth=2, label='Tração na corda (B)'),
    plt.Line2D([0], [0], color='purple', linewidth=3, label='Corda (4-5-3)'),
    plt.Line2D([0], [0], color='orange', linewidth=2, label='Momento em A')
]
ax.legend(handles=legend_elements, loc='upper right')

# Remover eixos
ax.set_xticks([])
ax.set_yticks([])

# Parede no ponto A
ax.axvline(x=-0.1, ymin=0.4, ymax=0.6, color='gray', linewidth=8)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Configuração da figura
fig, ax = plt.subplots(figsize=(10, 6))
ax.set_xlim(-1, 4)
ax.set_ylim(-1, 3)
ax.set_aspect('equal')
ax.grid(True, linestyle='--', alpha=0.7)
ax.set_title('Diagrama de Corpo Livre (DCL) - Força em A', fontsize=14, fontweight='bold')

# Desenhar a barra (horizontal de 0 a 3 m)
ax.plot([0, 3], [0, 0], 'k-', linewidth=3, label='Barra')

# Pontos de apoio A, B, C
points = [0, 2, 3]  # A=0m, B=2m, C=3m
labels = ['A', 'B', 'C']
for i, x in enumerate(points):
    ax.plot(x, 0, 'ro', markersize=8)
    ax.text(x, -0.3, labels[i], fontsize=12, ha='center')

# Força de 400 N a 30° APLICADA NO PONTO A (x=0)
force_x = 0  # Ponto A
force_angle = np.radians(30)
force_length = 1.0
dx = force_length * np.cos(force_angle)
dy = force_length * np.sin(force_angle)
ax.arrow(force_x, 0, dx, dy, head_width=0.1, head_length=0.1, fc='blue', ec='blue')
ax.text(force_x + dx/2 + 0.1, dy/2, '400 N', fontsize=10, color='blue', ha='left')

# Reações nos apoios (setas para cima)
reaction_length = 0.5
for x, label in zip(points, labels):
    ax.arrow(x, 0, 0, reaction_length, head_width=0.1, head_length=0.1, fc='red', ec='red')
    ax.text(x, reaction_length + 0.1, f'R_{label}', fontsize=10, color='red', ha='center')

# Reação horizontal em A (A_x) - agora deve equilibrar a componente horizontal da força
ax.arrow(0, 0, -reaction_length, 0, head_width=0.1, head_length=0.1, fc='green', ec='green')
ax.text(-reaction_length/2, 0.2, 'A_x', fontsize=10, color='green', ha='center')

# Momento em A (M_A) - para equilibrar o momento causado pela força
arc = plt.Circle((0, 0), 0.2, color='purple', fill=False, linestyle='--')
ax.add_patch(arc)
ax.text(0.3, 0.3, 'M_A', fontsize=10, color='purple')

# Indicar dimensões
ax.text(1, -0.5, '2 m', fontsize=10, ha='center')
ax.text(2.5, -0.5, '1 m', fontsize=10, ha='center')

# Legendas
ax.legend(loc='upper right')

# Remover eixos
ax.set_xticks([])
ax.set_yticks([])

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Circle, Arrow, Arc

# Configuração da figura
fig, ax = plt.subplots(figsize=(10, 8))
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)
ax.set_aspect('equal')
ax.grid(True, linestyle='--', alpha=0.7)
ax.axhline(y=0, color='k', linestyle='-', alpha=0.3)
ax.axvline(x=0, color='k', linestyle='-', alpha=0.3)
ax.set_title('Diagrama de Corpo Livre - Anel em A', fontsize=14, fontweight='bold')

# Ponto A (origem)
ax.plot(0, 0, 'ro', markersize=10, label='Anel em A')

# Força de 4N (horizontal para direita)
ax.arrow(0, 0, 2, 0, head_width=0.2, head_length=0.2, fc='blue', ec='blue')
ax.text(2.2, 0.1, '4N', fontsize=12, color='blue')

# Força de 3N (vertical para baixo)
ax.arrow(0, 0, 0, -1.5, head_width=0.2, head_length=0.2, fc='green', ec='green')
ax.text(0.3, -1.7, '3N', fontsize=12, color='green')

# Peso 200N (vertical para baixo)
ax.arrow(0, 0, 0, -3, head_width=0.2, head_length=0.2, fc='red', ec='red')
ax.text(0.3, -3.2, '200N', fontsize=12, color='red')

# Força do cabo AB (ângulo de 30° com a horizontal)
angle_ab = np.radians(30)
x_ab = 3 * np.cos(angle_ab)
y_ab = 3 * np.sin(angle_ab)
ax.arrow(0, 0, x_ab, y_ab, head_width=0.2, head_length=0.2, fc='purple', ec='purple')
ax.text(x_ab+0.3, y_ab+0.1, 'F_AB', fontsize=12, color='purple')

# Força do cabo AC (ângulo de 150° com a horizontal - assumindo para cima e esquerda)
angle_ac = np.radians(150)
x_ac = 3 * np.cos(angle_ac)
y_ac = 3 * np.sin(angle_ac)
ax.arrow(0, 0, x_ac, y_ac, head_width=0.2, head_length=0.2, fc='orange', ec='orange')
ax.text(x_ac+0.3, y_ac+0.1, 'F_AC', fontsize=12, color='orange')

# Ângulo de 30°
arc = Arc((0, 0), 1.5, 1.5, theta1=0, theta2=30, color='gray', linestyle='--')
ax.add_patch(arc)
ax.text(0.8, 0.3, '30°', fontsize=10, color='gray')

# Legenda
ax.legend(loc='upper right')

# Remover eixos
ax.set_xticks([])
ax.set_yticks([])

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define a função f(x, y) da EDO
def f(x, y):
    return y - x**2

# Cria a grade para o campo de direções
x_vals = np.linspace(-2, 4, 20)
y_vals = np.linspace(-2, 4, 20)
X, Y = np.meshgrid(x_vals, y_vals)
U = 1
V = f(X, Y)
N = np.sqrt(U**2 + V**2)
U = U / N
V = V / N

# Plota o campo de direções
plt.figure(figsize=(10, 8))
plt.quiver(X, Y, U, V, color='gray')

# Define a solução particular
def solucao_particular(x):
    return x**2 + 2*x + 2 - np.exp(x)

# Plota a curva da solução
x_sol = np.linspace(-2, 3.5, 100)
y_sol = solucao_particular(x_sol)
plt.plot(x_sol, y_sol, color='blue', linewidth=2, label='Curva Solução')

# Plota o ponto inicial
plt.plot(0, 1, 'ro', markersize=8, label='Ponto Inicial (0, 1)')
plt.text(0.1, 1.1, '(0, 1)', fontsize=12)

# Adiciona legendas e títulos
plt.title('Campo de Direções e Curva Solução')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.axis('equal') # Mantém a proporção dos eixos
plt.show()
```
