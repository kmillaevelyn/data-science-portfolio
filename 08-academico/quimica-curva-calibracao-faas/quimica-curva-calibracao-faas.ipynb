```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress

# Dados da curva analítica
concentracao_padrao = np.array([0, 0.5, 2.5, 5, 7.5, 10])
absorbancia_padrao = np.array([0.0012, 0.0278, 0.1348, 0.2765, 0.4171, 0.5601])

# Realizar a regressão linear
slope, intercept, r_value, p_value, std_err = linregress(concentracao_padrao, absorbancia_padrao)

# Equação da reta
# y = ax + b
# a = slope (coeficiente angular)
# b = intercept (coeficiente linear)
equacao_reta = f'y = {slope:.4f}x + {intercept:.4f}'
r_quadrado = r_value**2

# Gerar pontos para a linha de regressão
x_regressao = np.linspace(min(concentracao_padrao), max(concentracao_padrao), 100)
y_regressao = slope * x_regressao + intercept

# Plotar o gráfico
plt.figure(figsize=(10, 6))
plt.scatter(concentracao_padrao, absorbancia_padrao, color='blue', label='Dados Padrão')
plt.plot(x_regressao, y_regressao, color='red', label=f'Curva de Calibração: {equacao_reta}\nR² = {r_quadrado:.4f}')
plt.title('Curva de Calibração para a Determinação de Ferro por FAAS')
plt.xlabel('Concentração do Fe (mg/L)')
plt.ylabel('Absorbância')
plt.grid(True)
plt.legend()
plt.annotate(f'Coeficiente Angular (a) = {slope:.4f}', xy=(0.05, 0.95), xycoords='axes fraction')
plt.annotate(f'Coeficiente Linear (b) = {intercept:.4f}', xy=(0.05, 0.90), xycoords='axes fraction')
plt.show()

print(f"Equação da Reta da Curva de Calibração: y = {slope:.4f}x + {intercept:.4f}")
print(f"Coeficiente Angular (a): {slope:.4f}")
print(f"Coeficiente Linear (b): {intercept:.4f}")
print(f"Coeficiente de Determinação (R²): {r_quadrado:.4f}")
```
