```python
import streamlit as st
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Layout
st.set_page_config(layout="wide", page_title="Dashboard Interativo")

st.title("📊 Dashboard Interativo com Streamlit")

#Carregar dados com cache
@st.cache_data
def carregar_dados():
    return pd.read_csv("enhanced_student_habits_performance_dataset.csv")

df = carregar_dados()

#Verificar se há uma coluna de datas
coluna_data = None
for col in df.columns:
    if "data" in col.lower():
        coluna_data = col
        df[coluna_data] = pd.to_datetime(df[coluna_data])
        break

# Filtro por período (caso exista coluna de data)
if coluna_data:
    anos = df[coluna_data].dt.year.dropna().unique()
    ano_escolhido = st.sidebar.selectbox("Selecione o ano", sorted(anos))
    df_filtrado = df[df[coluna_data].dt.year == ano_escolhido]
else:
    st.sidebar.warning("Nenhuma coluna de data foi detectada.")
    df_filtrado = df

# Mostrar dados
st.subheader("📄 Tabela de Dados")
st.dataframe(df_filtrado, use_container_width=True)

# Gráficos
st.subheader("📈 Gráficos Interativos")

col1, col2 = st.columns(2)

with col1:
    st.markdown("### 📊 Gráfico de Linha")
    colunas_numericas = df_filtrado.select_dtypes(include='number').columns
    if len(colunas_numericas) >= 1 and coluna_data:
        y_col = st.selectbox("Selecione a variável para o eixo Y (linha)", colunas_numericas)
        fig, ax = plt.subplots()
        df_filtrado.groupby(coluna_data)[y_col].mean().plot(ax=ax)
        st.pyplot(fig)
    else:
        st.info("Não há dados numéricos ou coluna de data para gráfico de linha.")

with col2:
    st.markdown("### 📊 Gráfico de Barras")
    col_cat = st.selectbox("Selecione a coluna categórica (barras)", df_filtrado.select_dtypes(include='object').columns)
    fig, ax = plt.subplots()
    df_filtrado[col_cat].value_counts().plot(kind='bar', ax=ax)
    st.pyplot(fig)

col3, col4 = st.columns(2)

with col3:
    st.markdown("### 📊 Gráfico de Dispersão")
    if len(colunas_numericas) >= 2:
        x_disp = st.selectbox("X (dispersão)", colunas_numericas, index=0)
        y_disp = st.selectbox("Y (dispersão)", colunas_numericas, index=1)
        fig, ax = plt.subplots()
        sns.scatterplot(data=df_filtrado, x=x_disp, y=y_disp, ax=ax)
        st.pyplot(fig)
    else:
        st.info("São necessárias pelo menos 2 colunas numéricas para o gráfico de dispersão.")

with col4:
    st.markdown("### 📊 Gráfico de Pizza")
    if col_cat:
        fig, ax = plt.subplots()
        df_filtrado[col_cat].value_counts().plot(kind='pie', autopct='%1.1f%%', ax=ax)
        ax.set_ylabel("")
        st.pyplot(fig)
    else:
        st.info("Escolha uma coluna categórica para o gráfico de pizza.")
```
