```python
%%writefile app.py
import streamlit as st
import pandas as pd
import plotly.express as px

# Função para carregar os dados
@st.cache_data
def load_data():
    return pd.read_csv("dados.csv")

# Carregando os dados
df = load_data()

# Título do dashboard
st.title("📊 Análise de Estudantes e Saúde Acadêmica")

# Visualização da tabela
st.subheader("📋 Visualização da Tabela de Dados")
st.dataframe(df)

# Filtro de idade
st.sidebar.header("🔎 Filtros")
faixa_idade = st.sidebar.slider("Selecione a faixa etária:",
                                 int(df['age'].min()),
                                 int(df['age'].max()),
                                 (int(df['age'].min()), int(df['age'].max())))

df_filtrado = df[(df['age'] >= faixa_idade[0]) & (df['age'] <= faixa_idade[1])]

# Gráfico de dispersão: Horas de Estudo vs Nota
st.subheader("🎯 Relação entre Horas de Estudo por Dia e Nota Final")
fig1 = px.scatter(
    df_filtrado,
    x="study_hours_per_day",
    y="exam_score",
    color="gender",
    title="Horas de Estudo vs Nota Final",
    labels={"study_hours_per_day": "Horas de Estudo por Dia", "exam_score": "Nota Final"}
)
st.plotly_chart(fig1)

# Gráfico de barras: Nível de Ansiedade em Exames por Gênero
st.subheader("💥 Nível de Ansiedade em Exames por Gênero")
fig2 = px.box(
    df_filtrado,
    x="gender",
    y="exam_anxiety_score",
    color="gender",
    labels={"gender": "Gênero", "exam_anxiety_score": "Ansiedade em Exames"},
    title="Distribuição de Ansiedade em Exames por Gênero"
)
st.plotly_chart(fig2)

# Histograma: Horas de Sono
st.subheader("😴 Distribuição de Horas de Sono")
fig3 = px.histogram(
    df_filtrado,
    x="sleep_hours",
    nbins=20,
    title="Distribuição de Horas de Sono",
    labels={"sleep_hours": "Horas de Sono"}
)
st.plotly_chart(fig3)

# Gráfico de Pizza: Participação em Atividades Extracurriculares
st.subheader("🎭 Participação em Atividades Extracurriculares")
fig4 = px.pie(
    df_filtrado,
    names="extracurricular_participation",
    title="Participação em Atividades Extracurriculares"
)
st.plotly_chart(fig4)
```
