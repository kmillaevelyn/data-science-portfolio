```python
import numpy as np
import matplotlib.pyplot as plt

# Dados estimados (ano vs. população em milhares)
anos = np.array([2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023])
populacao = np.array([142.1, 144.3, 146.5, 148.7, 150.9, 153.1, 155.3, 157.5, 159.7, 161.9, 164.1])  # em milhares
```

```python
# Cálculo dos coeficientes
n = len(anos)
Sx = np.sum(anos)
Sy = np.sum(populacao)
Sxy = np.sum(anos * populacao)
Sxx = np.sum(anos ** 2)

a = (n * Sxy - Sx * Sy) / (n * Sxx - Sx ** 2)  # Coeficiente angular
b = (Sy - a * Sx) / n                           # Coeficiente linear

print(f"Equação da reta: y = {a:.4f}x + {b:.4f}")
```

```python
# Gerar pontos da reta ajustada
y_ajuste = a * anos + b

# Plotar dados reais e ajuste
plt.figure(figsize=(10, 6))
plt.scatter(anos, populacao, color='red', label='Dados Reais (Estimados)')
plt.plot(anos, y_ajuste, label=f'Ajuste Linear: y = {a:.2f}x + {b:.2f}', linestyle='--')
plt.title('Crescimento Populacional de Pouso Alegre (2013-2023)')
plt.xlabel('Ano')
plt.ylabel('População (milhares)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
ano_futuro = 2025
pop_futuro = a * ano_futuro + b
print(f"Previsão para 2025: {pop_futuro:.1f} mil habitantes")
```
