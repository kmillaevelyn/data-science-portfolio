```python
import collections
import heapq

# Define the graph of Romania as a dictionary of dictionaries
graph = {
    'Arad': {'Zerind': 75, 'Sibiu': 140, 'Timisoara': 118},
    'Zerind': {'Arad': 75, 'Oradea': 71},
    'Oradea': {'Zerind': 71, 'Sibiu': 151},
    'Sibiu': {'Arad': 140, 'Oradea': 151, 'Fagaras': 99, 'Rimnicu Vilcea': 80},
    'Timisoara': {'Arad': 118, 'Lugoj': 111},
    'Lugoj': {'Timisoara': 111, 'Mehadia': 70},
    'Mehadia': {'Lugoj': 70, 'Drobeta': 75},
    'Drobeta': {'Mehadia': 75, 'Craiova': 120},
    'Craiova': {'Drobeta': 120, 'Rimnicu Vilcea': 146, 'Pitesti': 138},
    'Rimnicu Vilcea': {'Sibiu': 80, 'Pitesti': 97, 'Craiova': 146},
    'Pitesti': {'Rimnicu Vilcea': 97, 'Craiova': 138, 'Bucareste': 101},
    'Fagaras': {'Sibiu': 99, 'Bucareste': 211},
    'Bucareste': {'Fagaras': 211, 'Pitesti': 101, 'Giurgiu': 90, 'Urziceni': 85},
    'Giurgiu': {'Bucareste': 90},
    'Urziceni': {'Bucareste': 85, 'Vaslui': 142, 'Hirsova': 98},
    'Hirsova': {'Urziceni': 98, 'Eforie': 86},
    'Eforie': {'Hirsova': 86},
    'Vaslui': {'Urziceni': 142, 'Iasi': 92},
    'Iasi': {'Vaslui': 92, 'Neamt': 87},
    'Neamt': {'Iasi': 87}
}

class Node:
    def __init__(self, state, parent, cost):
        self.state = state
        self.parent = parent
        self.cost = cost

def uniform_cost_search(start, goal):
    """
    Implements the Uniform Cost Search algorithm.
    """
    print("Iniciando a busca de custo uniforme...")
    frontier = []
    heapq.heappush(frontier, (0, start, [start]))
    explored = set()

    while frontier:
        cost, current_state, path = heapq.heappop(frontier)
        print(f"Nó expandido: {current_state} (Custo do caminho: {cost})")

        if current_state == goal:
            return path, cost

        if current_state in explored:
            continue

        explored.add(current_state)

        for neighbor, weight in graph[current_state].items():
            if neighbor not in explored:
                new_cost = cost + weight
                new_path = path + [neighbor]
                heapq.heappush(frontier, (new_cost, neighbor, new_path))
    return None, None

def breadth_first_search(start, goal):
    """
    Implements the Breadth-First Search algorithm.
    """
    print("Iniciando a busca em largura...")
    queue = collections.deque([[start]])
    explored = set([start])

    while queue:
        path = queue.popleft()
        current_state = path[-1]
        print(f"Nó expandido: {current_state}")

        if current_state == goal:
            return path

        for neighbor in graph[current_state]:
            if neighbor not in explored:
                explored.add(neighbor)
                new_path = path + [neighbor]
                queue.append(new_path)
    return None

def depth_first_search(start, goal):
    """
    Implements the Depth-First Search algorithm.
    """
    print("Iniciando a busca em profundidade...")
    stack = [(start, [start])]
    explored = set()

    while stack:
        current_state, path = stack.pop()

        if current_state not in explored:
            explored.add(current_state)
            print(f"Nó expandido: {current_state}")

            if current_state == goal:
                return path

            for neighbor in reversed(list(graph[current_state].keys())):
                if neighbor not in explored:
                    new_path = path + [neighbor]
                    stack.append((neighbor, new_path))
    return None

# Problem 1: Uniform Cost Search
print("--- Algoritmo de Busca de Custo Uniforme ---")
ucs_path, ucs_cost = uniform_cost_search('Arad', 'Bucareste')
if ucs_path:
    print("\nCidades percorridas no caminho:")
    for city in ucs_path:
        print(city)
    print(f"\nCusto total do caminho: {ucs_cost}")
else:
    print("Não foi possível encontrar um caminho.")

print("\n" + "="*50 + "\n")

# Problem 2: Breadth-First Search
print("--- Algoritmo de Busca em Largura ---")
bfs_path = breadth_first_search('Arad', 'Bucareste')
if bfs_path:
    print("\nCidades percorridas no processo de busca:")
    for city in bfs_path:
        print(city)
else:
    print("Não foi possível encontrar um caminho.")

print("\n" + "="*50 + "\n")

# Problem 3: Depth-First Search
print("--- Algoritmo de Busca em Profundidade ---")
dfs_path = depth_first_search('Arad', 'Bucareste')
if dfs_path:
    print("\nCidades percorridas no processo de busca:")
    for city in dfs_path:
        print(city)
else:
    print("Não foi possível encontrar um caminho.")
```
