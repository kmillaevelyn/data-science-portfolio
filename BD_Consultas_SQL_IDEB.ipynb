```python
#Questão 11

import mysql.connector
from mysql.connector import errorcode

conexao = mysql.connector.connect(
  host='test_db.mysql.dbaas.com.br',
   user='test_db',
   password='BD@2024bctec',
   database='test_db'
)

cursor = conexao.cursor()

comando = 'select ano, ideb from ideb where ideb = (select MAX(ideb) from ideb);'

cursor.execute(comando)
resultado = cursor.fetchone()  # Obtém o resultado em uma única linha

if resultado:
    ano_maior_ideb, maior_ideb = resultado
    print(f'O ano com o maior valor para o IDEB foi: {ano_maior_ideb} (IDEB: {maior_ideb})')
else:
    print('Nenhum dado encontrado na tabela ideb.')
```

```python
#Questão 12

import mysql.connector
from mysql.connector import errorcode

conexao = mysql.connector.connect(
   host='test_db.mysql.dbaas.com.br',
   user='test_db',
   password='BD@2024bctec',
   database='test_db'
)

cursor = conexao.cursor()

comando = 'select rede, MAX(projecao) AS maior_projecao from ideb GROUP BY projecao'

cursor.execute(comando)

resultados = cursor.fetchall()

print("Maiores valores de 'projecao'")

for resultado in resultados:
    tipo_rede = resultado[0]
    maior_projecao = resultado[1]

    print(f"Tipo de Rede: {tipo_rede}, Maior Projecao: {maior_projecao}")

conexao.close()
```
