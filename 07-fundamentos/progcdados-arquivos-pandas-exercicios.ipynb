```python
import csv
#questão 3
# Dados dos produtos
produtos = [
    {"ID": 1, "Produto": "Açúcar", "Preço": 10.53},
    {"ID": 2, "Produto": "Café", "Preço": 20.79},
    {"ID": 3, "Produto": "Pão de forma", "Preço": 15.32}
]

with open('produtos.csv', 'w', newline='', encoding='utf-8') as arquivo_csv:
    campos = ["ID", "Produto", "Preço"]
    escritor = csv.DictWriter(arquivo_csv, fieldnames=campos)

    # Escrevendo o cabeçalho
    escritor.writeheader()

    # Escrevendo os dados dos produtos
    escritor.writerows(produtos)

print("Arquivo 'produtos.csv' criado com sucesso!")

#questão 4
# Mostrando o conteúdo criado
print("\nConteúdo do arquivo:")
with open('produtos.csv', 'r', encoding='utf-8') as arquivo:
    for linha in arquivo:
        print(linha.strip())
```

```python
#questão 5
produtos = []
with open('produtos.csv', 'r', encoding='utf-8') as arquivo_original:
    leitor = csv.DictReader(arquivo_original)
    for produto in leitor:
        produtos.append(produto)

# Aplicando 10% de desconto nos preços
for produto in produtos:
    preco_original = float(produto['Preço'])
    produto['Preço'] = round(preco_original * 0.9, 2)  # 10% de desconto

# Criando o novo arquivo com descontos
with open('produtos_desconto.csv', 'w', newline='', encoding='utf-8') as arquivo_desconto:
    campos = ["ID", "Produto", "Preço"]
    escritor = csv.DictWriter(arquivo_desconto, fieldnames=campos)

    escritor.writeheader()
    escritor.writerows(produtos)

print("Arquivo 'produtos_desconto.csv' criado com sucesso!")

# Mostrando o conteúdo do novo arquivo
print("\nProdutos com desconto de 10%:")
with open('produtos_desconto.csv', 'r', encoding='utf-8') as arquivo:
    for linha in arquivo:
        print(linha.strip())
```

```python
#questão 6
import pandas as pd
dados = {
    "Aluno": ["Kamilla", "Guilherme", "Eliot", "Julya"],
    "Nota 1": [7.5, 8.2, 6.9, 9.1],
    "Nota 2": [8.0, 7.5, 6.5, 9.5]
}

df = pd.DataFrame(dados)

# 1. Exibindo apenas a coluna Aluno
print("\n1. Coluna 'Aluno':")
print(df['Aluno'])

# 2. Exibindo apenas a primeira linha do DataFrame
print("\n2. Primeira linha do DataFrame:")
print(df.iloc[0])

# 3. Calculando a média das notas para o aluno José
media_Kamilla = (df.loc[df['Aluno'] == 'Kamilla', 'Nota 1'].values[0] +
              df.loc[df['Aluno'] == 'Kamilla', 'Nota 2'].values[0]) / 2
print("\n3. Média das notas do Kamilla:")
print(media_Kamilla)

# Versão alternativa usando mean() para calcular a média
print("\n3a. Média (método alternativo):")
print(df[df['Aluno'] == 'Kamilla'][['Nota 1', 'Nota 2']].mean(axis=1).values[0])
```

```python
#questão 7
dados = {
    "Nome": ["Guilherme", "Kamilla", "Julya", "Eliot"],
    "Idade": [25, 35, 28, 45],
    "Salário": [2500, 3500, 2800, 4200]
}

df = pd.DataFrame(dados)

# 1. Filtro para idade > 30
empregados_idade_maior_30 = df[df['Idade'] > 30]
print("\nEmpregados com idade maior que 30:")
print(empregados_idade_maior_30)

# 2. Filtro para salário > 3000
empregados_salario_maior_3000 = df[df['Salário'] > 3000]
print("\nEmpregados com salário maior que 3000:")
print(empregados_salario_maior_3000)

# 3. Gravando o DataFrame em um arquivo CSV
df.to_csv('empregados.csv', index=False, encoding='utf-8')
print("\nDataFrame gravado no arquivo 'empregados.csv'")

# 4. Lendo o arquivo CSV
df_lido = pd.read_csv('empregados.csv')
print("\nConteúdo do arquivo CSV lido:")
print(df_lido)
```
