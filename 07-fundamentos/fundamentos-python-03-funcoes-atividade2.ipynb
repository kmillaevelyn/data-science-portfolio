```python
# Exercício 3 -  Visualizando um gráfico de equação quadrativa

print("Composiçao da equação de 2° = ax²+bx+c=0")

print("Determine os valores")  # coeficientes da equação quadratica
a = int(input("Informe o valor de a: "))
b = int(input("Informe o valor de b: "))
c = int(input("Informe o valor de c: "))

delta = (b**2) - (4 * a * c)


# Formula de Bhaskara
x1 = (- b + (delta ** (1/2))) / 2 * a
x2 = (- b - (delta ** (1/2))) / 2 * a

# Resultados
if x1 == x2:
  print(f"O resultado de x1 e x2 são iguais a {x1}")
else:
  print(f"O resultado de x1 é igual a {x1} e x2 é igual a {x2}")

# Resultados para plotagem do gráfico
xval = []
yval = []

for x in range (-8,10,1):
  xval.append(x)
  yval.append((a * (x ** 2)) + (b * x) +c)

print("Valores de x: ", xval)
print("Valores de y: ", yval)


# Importando para o gráfico
x = np.linspace(-10, 10, 400)

#x = np.linspace(campo1, campo2, campo2)


plt.plot(xval, yval, label=f'{a}x^2 + {b}x + {c}', color='blue')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Grafico 2°')
plt.legend()
plt.show()
```
