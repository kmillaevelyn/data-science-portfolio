```python
# Solicitar dois números do usuário
numero1 = float(input("Digite o primeiro número: "))
numero2 = float(input("Digite o segundo número: "))

# Realizar operações básicas
soma = numero1 + numero2
subtracao = numero1 - numero2
multiplicacao = numero1 * numero2
divisao = numero1 / numero2

# Exibir os resultados
print("Soma:", soma)
print("Subtração:", subtracao)
print("Multiplicação:", multiplicacao)
print("Divisão:", divisao)
```

```python
# Exercício 1 - Calculo de IMC em python
print("Vamos calcular seu IMC")
print(" "*30)
peso = float(input("Informe seu peso: "))
altura = float(input("Informe sua altura com ponto: "))
print("-"*30)

# Calculo para resultado do IMC
calculo = peso / altura ** 2

# Resultado
print("Seu IMC é: ", calculo)
```

```python
# Exercício 2 Convertendo celcius em fahrenheit
print("Vamos converter graus °C em °F")
print(" "*30)
graus = float(input("Informe a temperatura em graus celcius: "))
resultado = graus * (9 / 5) + 32

print('Resultado em Fihrenheit: {0}°F'.format(resultado))
```

```python
# Exercício 3 Fibonnaci

print("Sequencia de Fibonacci")
print(" "*30)
n = int(input("Quantos números da sequencia você quer mostrar? "))
t1 = 0
t2 = 1
print("="*30)
print("{} {}" .format(t1, t2), end="")
cont = 3
while cont <= n:
  t3 = t1 + t2
  print(" {}" .format(t3), end="")
  t1 = t2
  t2 = t3
  cont += 1
print(" Fim ")
```
