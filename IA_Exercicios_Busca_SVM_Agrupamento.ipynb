```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, dendrogram

# 1. PREPARAÇÃO DOS DADOS
# Os dados da tabela Pessoas: Cabelo, Peso, Idade
dados = {
    'Marcos': [0, 113, 36],
    'Ana': [10, 68, 34],
    'Ronaldo': [41, 2, 10],
    'Amanda': [6, 35, 8],
    'Marta': [4, 9, 1],
    'Josias': [77, 1, 70],
    'Rose': [8, 73, 41],
    'Claudio': [10, 82, 38],
    'Silvio': [6, 91, 45]
}

# Converte para DataFrame e Transpõe (Pessoas nas linhas)
df = pd.DataFrame(dados).T
df.columns = ['Cabelo', 'Peso', 'Idade']

# Rótulos para o dendrograma
rotulos = df.index.tolist()
dados_array = df.values

# 2. CÁLCULO DA LIGAÇÃO (LINKAGE)
# method='single': Ligação Simples (vizinho mais próximo)
# metric='euclidean': Distância Euclidiana
Z = linkage(dados_array, method='single', metric='euclidean')

# 3. GERAÇÃO E VISUALIZAÇÃO DO DENDROGRAMA
plt.figure(figsize=(12, 6))
plt.title('Dendrograma - Agrupamento Hierárquico (Ligação Simples)')
plt.xlabel('Pessoa')
plt.ylabel('Distância Euclidiana')

dendrogram(
    Z,
    orientation='top', # 'top' para o dendrograma na vertical
    labels=rotulos,
    leaf_rotation=45, # Rotação dos rótulos para melhor visualização
    leaf_font_size=10,
)

# Adiciona linhas horizontais nas alturas de corte para 2 e 3 clusters (opcional)
# Corte para 3 clusters (entre 32.14 e 60.15)
plt.axhline(y=35, color='r', linestyle='--', label='Corte para 3 Clusters')
# Corte para 2 clusters (entre 60.15 e o próximo)
plt.axhline(y=65, color='g', linestyle='-.', label='Corte para 2 Clusters')
plt.legend()

plt.grid(axis='y')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Definindo os pontos
pontos = {
    'A': [(2, 3), (3, 2), (4, 4), (4, 2)],
    'B': [(6, 4), (6, 3), (7, 2), (8, 3)]
}

# Criando o gráfico
plt.figure(figsize=(10, 8))

# Plotando os pontos da classe A
for ponto in pontos['A']:
    plt.scatter(ponto[0], ponto[1], c='blue', s=100, label='Classe A' if ponto == pontos['A'][0] else "")
    plt.text(ponto[0] + 0.1, ponto[1] + 0.1, f'({ponto[0]},{ponto[1]})', fontsize=9)

# Plotando os pontos da classe B
for ponto in pontos['B']:
    plt.scatter(ponto[0], ponto[1], c='red', s=100, label='Classe B' if ponto == pontos['B'][0] else "")
    plt.text(ponto[0] + 0.1, ponto[1] + 0.1, f'({ponto[0]},{ponto[1]})', fontsize=9)

# Desenhando o separador linear (SVM)
x_separador = np.array([4.5, 4.5])  # Linha vertical em x = 4.5
y_separador = np.array([1.5, 4.5])
plt.plot(x_separador, y_separador, 'g--', linewidth=2, label='Separador SVM')

# Destacando os vetores de suporte (pontos mais próximos do separador)
vetores_suporte = [(4, 2), (6, 3)]
for vs in vetores_suporte:
    plt.scatter(vs[0], vs[1], c='yellow', s=150, edgecolors='black', linewidth=2,
                label='Vetor Suporte' if vs == vetores_suporte[0] else "")

# Configurações do gráfico
plt.xlim(1, 9)
plt.ylim(1, 5)
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('Questão 8 - SVM com Separador Linear e Vetores de Suporte')
plt.grid(True, alpha=0.3)
plt.legend()
plt.gca().set_aspect('equal', adjustable='box')

plt.tight_layout()
plt.show()

# Imprimindo informações
print("=== QUESTÃO 8 ===")
print("\nPontos Classe A:", pontos['A'])
print("Pontos Classe B:", pontos['B'])
print("\nVetores de Suporte:", vetores_suporte)
print("Separador SVM: x ≈ 4.5 (linha verde tracejada)")
```
